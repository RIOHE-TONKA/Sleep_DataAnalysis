```python
#A) Change all the times in the ‘Duration’ column to integer minutes only 
# (e.g., ‘00:30’ becomes 30 and ‘01:30’ becomes 90) and add this as a new column called ‘Calculated’ in the dataframe.
import re
pattern = r'\d{1,2}:\d{1,2}$'


# def to_min(times):
#     if re.match(pattern, times) or re.match(pattern2, times) :
#         hour = int(times[:2])
#         min = int(times[-2:])

#         minutes = (60*hour)+min

#         return minutes
#     else: 
#         return False


pattern = r'^(\d{1,2}):(\d{1,2})$'

def to_min(times):
    match = re.match(pattern, times)
    if match:
        hour = int(match.group(1))  # first capture group = hour
        minute = int(match.group(2))  # second capture group = minute
        return hour * 60 + minute
    else:
        return False
    

print(to_min('30:30'))


data5['Calculated'] = data5['DURATION'].apply(lambda x: to_min(x))
```
